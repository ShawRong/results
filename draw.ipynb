```python
import os
def list_files_in_directory(directory_path):
    # 获取目录下的所有文件和文件夹
    files = os.listdir(directory_path)
    # 过滤出文件名（排除子目录）
    return [f for f in files if os.path.isfile(os.path.join(directory_path, f))]


def get_score(specify):
    path = './results/score/' + specify + '/'
    to_draw = list_files_in_directory(path)
    file2metrics = {} 
    for draw in to_draw:
        file_name = path + draw 
        metrics = {} 
        with open(file_name, 'r') as file:
            for line in file:
                # 去掉行首尾的空白字符
                line = line.strip()
                if line.startswith("IoU:"):
                    # 提取 IoU 值
                    metrics['IoU'] = float(line.split(":")[1].strip())
                elif line.startswith("Cor:"):
                    # 提取 Cor 值
                    metrics['Cor'] = float(line.split(":")[1].strip())
        file2metrics[draw] = metrics
    return file2metrics

specify = ['normal_score', 'k4_score', 'k5_score']
scores = []
for item in specify:
    scores.append(get_score(item))
scores = list(zip(specify, scores))
normal = scores[0][1]
k4 = scores[1][1]
k5 = scores[2][1]

        
```

```python
import pandas as pd
normal = pd.DataFrame(normal)
normal_iou = normal.iloc[[0]].reset_index(drop=True)  # 第一行
normal_cor = normal.iloc[[1]].reset_index(drop=True)  # 第一行
k4 = pd.DataFrame(k4)
k4_iou = k4.iloc[[0]].reset_index(drop=True)  # 第一行
k4_cor = k4.iloc[[1]].reset_index(drop=True)  # 第一行
k5 = pd.DataFrame(k5)
k5_iou = k5.iloc[[0]].reset_index(drop=True)  # 第一行
k5_cor = k5.iloc[[1]].reset_index(drop=True)  # 第一行
```

```python

iou = pd.concat([normal_iou, k4_iou, k5_iou], ignore_index=True, )
iou.index = ['k3', 'k4', 'k5']
iou
```

```python

cor = pd.concat([normal_cor, k4_cor, k5_cor], ignore_index=True, )
cor.index = ['k3', 'k4', 'k5']
cor
```
