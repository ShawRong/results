```python
import os
def list_files_in_directory(directory_path):
    # 获取目录下的所有文件和文件夹
    files = os.listdir(directory_path)
    # 过滤出文件名（排除子目录）
    return [f for f in files if os.path.isfile(os.path.join(directory_path, f))]


def get_score(specify):
    path = './results/score/' + specify + '/'
    to_draw = list_files_in_directory(path)
    file2metrics = {} 
    for draw in to_draw:
        file_name = path + draw 
        metrics = {} 
        with open(file_name, 'r') as file:
            for line in file:
                # 去掉行首尾的空白字符
                line = line.strip()
                if line.startswith("IoU:"):
                    # 提取 IoU 值
                    metrics['IoU'] = float(line.split(":")[1].strip())
                elif line.startswith("Cor:"):
                    # 提取 Cor 值
                    metrics['Cor'] = float(line.split(":")[1].strip())
        file2metrics[draw] = metrics
    return file2metrics

specify = ['normal_score', 'tmp0_score', 'tmp2_score', 'tmp3_score']
scores = []
for item in specify:
    scores.append(get_score(item))
scores = list(zip(specify, scores))
normal = scores[0][1]
t0 = scores[1][1]
t2 = scores[2][1]
t3 = scores[3][1]

        
```

```python
import pandas as pd
normal = pd.DataFrame(normal)
normal_iou = normal.iloc[[0]].reset_index(drop=True)  # 第一行
normal_cor = normal.iloc[[1]].reset_index(drop=True)  # 第一行
t0 = pd.DataFrame(t0)
t0_iou = t0.iloc[[0]].reset_index(drop=True)  # 第一行
t0_cor = t0.iloc[[1]].reset_index(drop=True)  # 第一行
t2 = pd.DataFrame(t2)
t2_iou = t2.iloc[[0]].reset_index(drop=True)  # 第一行
t2_cor = t2.iloc[[1]].reset_index(drop=True)  # 第一行
t3 = pd.DataFrame(t3)
t3_iou = t3.iloc[[0]].reset_index(drop=True)  # 第一行
t3_cor = t3.iloc[[1]].reset_index(drop=True)  # 第一行
```

```python

iou = pd.concat([t0_iou, normal_iou, t2_iou, t3_iou], ignore_index=True, )
iou.index = ['t0', 't1(base)', 't2', 't3']
iou
```

```python

cor = pd.concat([t0_cor, normal_cor, t2_cor, t3_cor], ignore_index=True, )
cor.index = ['t0', 't1(base)', 't2', 't3']
cor
```

```python
iou = iou.transpose()
iou
```

```python
cor = cor.transpose()
cor
```

```python
iou = iou.sort_values(by='t0')
iou
```

```python
import matplotlib.pyplot as plt
import numpy as np

df = iou[8:]

# 设置绘图
plt.figure(figsize=(10, 6))

colors = plt.cm.viridis(np.linspace(0, 1, 5))  # 生成17种颜色
markers = ['o', 's', '^', 'D', 'p', '*', 'x', 'h', 'v', '<', '>', '1', '2', '3', '4', '8']

# 为每一行绘制一条曲线
i = 0
for index, row in df.iterrows():
    plt.plot(df.columns, row, marker=markers[i % len(markers)], linestyle='-', label=index)
    i = i + 1

# 添加标题和标签
plt.title('Model Temperature t & IoU score (2)')
plt.xlabel('Model Temperature')
plt.ylabel('IoU', rotation = 90)
plt.grid()

# 显示图例
plt.legend(title='Lang&Model', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# 显示图形
plt.show()
```
